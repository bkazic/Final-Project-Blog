# Statistical tests

#### TWhelch's test: Speed at Day VS. Speed at Night

```python
import scipy.stats

# Add Hour and DayOrNight coulms
traffic_data["Hour"] = traffic_data.index.hour # Could I do this somehow without creating new column Hour
traffic_data["DayOrNight"] = traffic_data.Hour.apply(lambda x: 1 if (5 < x < 18) else 0)

# Split traffic_data in two dataframes -- traffic_data_day and traffic_data_night
traffic_data_day = traffic_data[traffic_data["DayOrNight"]==1]
traffic_data_night = traffic_data[traffic_data["DayOrNight"]==0]
traffic_data.drop(['Hour', 'DayOrNight'], axis=1, inplace=True)

# Perform Welch's t-test
pTest = scipy.stats.ttest_ind(traffic_data_day['Speed'], traffic_data_night['Speed'], equal_var=False)
means = (traffic_data_day['Speed'].mean(), traffic_data_night['Speed'].mean())
count = traffic_data_day.Speed.count(), traffic_data_night.Speed.count()

# If p value is larger than 5%, then we cannot reject the null hypothesis of identical average scores
if pTest[1] <= .05:
    result = ({"NullHypothesis": False}, {"WelchTest": pTest}, {"Means": means}, {"Count": count})
else:
    result = ({"NullHypothesis": True}, {"WelchTest": pTest}, {"Means": means}, {"Count": count})   
        
print "Speed by Day VS. Speed by Night"
result
```

# Modelling & Predictions

```python
import numpy as np
import pandas as pd
from scipy import linalg
import statsmodels # Is this used??

def predictions(dataframe):
    features = dataframe[["Flow","Gap","Occupancy","Speed","TrafficStatus"]]
    values = dataframe[["Target"]]

    features["ones"] = np.ones(len(values))
    features_array = np.array(features)
    values_array = np.array(values).flatten()
        
    theta_ols = linalg.lstsq(features_array, values_array)[0] 
    predictions = np.dot(features_array, theta_ols)
  
    return predictions

def compute_r_squared(data, predictions):
    SST = ((data-np.mean(data))**2).sum()
    SSReg = ((predictions-np.mean(data))**2).sum()
    r_squared = SSReg / SST

    return r_squared


traffic_data["Predictions"] = predictions(traffic_data)
print compute_r_squared(traffic_data["Target"], traffic_data["Predictions"])
#traffic_data
```
